```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("chess - chess tempo openings database.csv")
```

```python
df
```

```python
moves_list = df['Moves']
```

```python
df['moves_list']=moves_list.str.split(" ")
```

```python
df['move1w'] = df['moves_list'].apply(lambda x : x[0] )
```

```python
df['move1b'] = df['moves_list'].apply(lambda x : x[1] if len(x)>=2 else "" )
```

```python
df['move2w'] = df['moves_list'].apply(lambda x : x[2] if len(x)>=3 else "" )
```

```python
df['move2b'] = df['moves_list'].apply(lambda x : x[3] if len(x)>=4 else "" )
```

```python
df['move3w'] = df['moves_list'].apply(lambda x : x[4] if len(x)>=5 else "" )
df['move3b'] = df['moves_list'].apply(lambda x : x[5] if len(x)>=6 else "" )
df['move4w'] = df['moves_list'].apply(lambda x : x[6] if len(x)>=7 else "" )
df['move4b'] = df['moves_list'].apply(lambda x : x[7] if len(x)>=8 else "" )
```

```python
df['Colour'] = df['moves_list'].apply(lambda x : "white" if len(x)%2==1 else "black")
```

```python
 float(df["Player Win %"][0][:-1])
```

```python
df["Player Win %"] = df["Player Win %"].apply(lambda x : float(x[:-1] ))
```

```python
df["Opponent Win %"] = df["Opponent Win %"].apply(lambda x : float(x[:-1]) )
df["Draw %"] = df["Draw %"].apply(lambda x : float(x[:-1] ))
```

```python
df['White_win%'] = np.zeros
df['Colour'][1]
```

```python
for x in range(len(df['White_win%'])):
    if df['Colour'][x] == 'white':
        df['White_win%'][x] = df['Player Win %'][x]
    else:
        df['White_win%'][x] = df['Opponent Win %'][x]
```

```python
df
```

```python
df['Black_win%'] = np.zeros
for x in range(len(df['Black_win%'])):
    if df['Colour'][x] == 'black':
        df['Black_win%'][x] = df['Player Win %'][x]
    else:
        df['Black_win%'][x] = df['Opponent Win %'][x]
```

```python
df['White_odds'] = (df['White_win%']/df['Black_win%'])

```

```python
df['move1w'] = df['move1w'].str.replace('1.', '')
df['move2w'] = df['move2w'].str.replace('2.', '')
df['move3w'] = df['move3w'].str.replace('3.', '')
df['move4w'] = df['move4w'].str.replace('4.', '')
```

```python
df
```

```python
df.to_csv('chessdb0-1')
```
