```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('chessdb0-1')
```

```python
e4 = df[df['move1w']=='e4'].groupby('move1b').agg({'Num Games': np.sum, 'White_Wins' : np.sum, 'Black_Wins':np.sum})
```

```python
e4['White_odds'] = e4['White_Wins'] / e4['Black_Wins']
```

```python
e4.sort_values('White_odds')
```
